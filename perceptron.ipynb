```python
import numpy as np
import pandas as pd
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import to_categorical
```

```python
df = pd.read_csv('./DataSet/hw03_images.csv')
df_label = pd.read_csv('./DataSet/hw03_labels.csv')
df.shape
df_train = df.iloc[:500]
df_test = df.iloc[500:]
```

```python
df_label
```

```python
def encoder(lbl, cls):
    encoded = np.zeros(shape=(lbl.shape[0],cls))
    c = 0
    print(lbl.shape)
    for i in lbl['1']:
        
        iteration = int(i)
        encoded[c][iteration-1] = 1
        c = c+1
    print(encoded, encoded.shape)
    return encoded
encoded_train = encoder(df_label.iloc[:500], 5)
encoded_test = encoder(df_label.iloc[500:], 5)
```

```python
df_train.shape
```

```python
keras.optimizers.rmsprop(lr=0.0001)
df_test.shape
```

```python
model = Sequential()
model.add(Dense(20, activation='sigmoid', input_shape=(784,)))
model.add(Dense(5, activation='softmax'))
model.compile(optimizer='rmsprop', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
history = model.fit(df_train, encoded_train, batch_size=256, epochs=500, verbose=1, 
                   validation_data=(df_test, encoded_test))
```

```python
[test_loss, test_acc] = model.evaluate(df_test, encoded_test)
print("Evaluation result on Test Data : Loss = {}, accuracy = {}".format(test_loss, test_acc))
```
